# OBJECTIVE

Practice mocdel selection and evaluation using cross validation on a simulated dataset that represents a target variable representing the volume of traffic in cars/hour across a city bbridge and various normalized features related to traffic data such as the tme of the day.

# GOAL
To build a deep neural netwok model that predicts the traffic volume across the given various features

#DATA AND TYPE OF PROBLEM
The dataset contain 10000 records and for each of them,10 features are included.The target column is a countinous variable,thus It is a regression problem. 

```python
# libraries



import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,ELU,ReLU
```

```python
# loading the dataset into pandas dataframe object

feature=pd.read_csv("/content/traffic_volume_feats.csv")
target=pd.read_csv("/content/traffic_volume_target.csv")
```

```python
# define three function, each returning keras model and using cross val to pick the best out of the three

def model1():
  model=Sequential()
  model.add(Dense(input_dim=feature.shape[1],units=10 ,activation="relu"))
  model.add(Dense(units=1 ,activation="relu"))

  model.compile(loss=["mean_squared_error"],metrics="mean_squared_error",optimizer="adam")

  return model
  
def model2():
  model=Sequential()
  model.add(Dense(input_dim=feature.shape[1],units=10 ,activation="relu"))
  model.add(Dense(units=10 ,activation="relu"))
  model.add(Dense(units=1 ,activation="relu"))

  model.compile(loss=["mean_squared_error"],metrics="mean_squared_error",optimizer="adam")
  return model

def model3():
  model=Sequential()
  model.add(Dense(input_dim=feature.shape[1],units=10 ,activation="relu"))
  model.add(Dense(units=10 ,activation="relu"))
  model.add(Dense(units=10 ,activation="relu"))
  model.add(Dense(units=1 ,activation="relu"))

  model.compile(loss=["mean_squared_error"],metrics="mean_squared_error",optimizer="adam")
  return model



```

```python
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold,cross_val_score

results=[]
models=[model1,model2,model3]
kf=KFold(5)
for model in models:
  my_model=KerasRegressor(model,epochs=100,shuffle=False,batch_size=5)
  result=cross_val_score(my_model,feature,target,cv=kf)
  results.append(abs(result).mean())
```

```python
print(results)
```

```python
# the model with the lowest loss

print(f"THE MODEL WITH THE LEAST ERROR IS MODEL{results.index(min(results))+1} WITH ERROR OF {min(results)}")
```

```python
epochs=[80,100]
batch=[50,25]
kf=KFold(5)
results=[]
for i in epochs:
  for j in batch:
    my_model=KerasRegressor(model1,epochs=i,batch_size=j)
    result=cross_val_score(my_model,feature,target,cv=kf)
    results.append({"epoch":i,"batch_size":j,"mean_error":abs(result).mean()})
```

```python
# the best batch size is 50 and the best epoch is 80
```

```python

opt=["rmsprop","adam","sgd"]
results=[]
kf=KFold(5)
for i in opt:
  def model2():
    model=Sequential()
    model.add(Dense(input_dim=feature.shape[1],units=10 ,activation="relu"))
    model.add(Dense(units=10 ,activation="relu"))
    model.add(Dense(units=1 ,activation="relu"))

    model.compile(loss=["mean_squared_error"],metrics="mean_squared_error",optimizer=i)
    return model

    
  my_model=KerasRegressor(model,epochs=80,batch_size=50)
  result=cross_val_score(my_model,feature,target,cv=kf)
  results.append({"optimizer":i,"mean_error":abs(result).mean()})
```
