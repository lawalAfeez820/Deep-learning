# OBJECTIVE

To use Keras Sequential model to build a binary classifier.The dataset provided represent the testing results of the production of aircraft propellers. The target variable is the results  of the manual inspection of propellers,designated as either pass (1) or fail(0).

# Goal

The goal is to classify the testing  results into either pass or fail classes to math the manual inpections .The performance of the model will be obsered visually using matplotlib,This helps in gaining better sense of how going from one processing unit of layer of processing units changes the flexibility and performance of the model

#  Data 
The dataset contain two features representing the test results of two different tests inspecting the aircraft propellers of over 3000 propellers.

# Business Objective
Developing an automated test to access the propeller faster ,therefore the goal is to build a model that can predict whether an aircraft propeller  will pass the manual inspection when given the results of the two tests.

A model with o hidden layer(logistic regression) was first built, then A single hidden layer with three and six nodes were built to perform the classification.

```python
# Importing the libraries

import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,ELU,ReLU
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,confusion_matrix,accuracy_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading the dataset

feature=pd.read_csv("/content/outlier_feats.csv")
target=pd.read_csv("/content/outlier_target.csv")
```

```python
#reading the dataset

np.unique(target.values)
```

```python
feature.head()
```

```python
#spliting the dataset into train and testdatset

x_train,x_test,y_train,y_test=train_test_split(feature,target,stratify=target,random_state=100)
```

```python
# Buiding a model with no hidden layer using sigmoid as  the  activation function 

model1=Sequential()
model1.add(Dense(input_dim=x_train.shape[1],units=1,activation="sigmoid"))

model1.compile(metrics=["accuracy"],optimizer="sgd",loss="binary_crossentropy")
```

```python
# fitting the model on the train data with 100 epochs

history1=model1.fit(x_train,y_train,batch_size=5,epochs=100,validation_split=0.2)
```

the loss on validation set after 100 epochs is 0.3537

```python
#ploting the accuracy of the train and val data against epoch

plt.plot(history1.history["accuracy"])
plt.plot(history1.history["val_accuracy"])
plt.title("Model accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train","Validation"])
plt.show()
```

```python
#ploting the loss of the train and val data against epoch

plt.plot(history1.history["loss"])
plt.plot(history1.history["val_loss"])
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.legend(["Train","Validation"])
plt.show()
```

```python
# Buiding a model with one hidden layer with 3 nodes using sigmoid as the  activation function of output layer

model2=Sequential()
model2.add(Dense(input_dim=x_train.shape[1],units=3,activation="relu"))
model2.add(Dense(units=1,activation="sigmoid"))

model2.compile(metrics=["accuracy"],optimizer="sgd",loss="binary_crossentropy")
```

```python
# fitting the model on the train data with 200 epochs

history2=model2.fit(x_train,y_train,batch_size=5,epochs=100,validation_split=0.2)
```

The loss of the validation set after 200 epochs is 0.1509

```python
#ploting the accuracy of the train and val data against epoch

plt.plot(history2.history["accuracy"])
plt.plot(history2.history["val_accuracy"])
plt.title("Model accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train","Validation"])
plt.show()
```

```python
#ploting the loss of the train and val data against epoch

plt.plot(history2.history["loss"])
plt.plot(history2.history["val_loss"])
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.legend(["Train","Validation"])
plt.show()
```

```python
# Buiding a model with one hidden layer with 6 nodes using sigmoid as the  activation function of output layer

model3=Sequential()
model3.add(Dense(input_dim=x_train.shape[1],units=6,activation="relu"))
model3.add(Dense(units=1,activation="sigmoid"))

model3.compile(metrics=["accuracy"],optimizer="sgd",loss="binary_crossentropy")
```

```python
# fitting the model on the train data with 400 epochs

history3=model2.fit(x_train,y_train,batch_size=5,epochs=400,validation_split=0.2)
```

The loss on the validation dataset has decreased


```python
#ploting the accuracy of the train and val data against epoch

plt.plot(history3.history["accuracy"])
plt.plot(history3.history["val_accuracy"])
plt.title("Model accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train","Validation"])
plt.show()
```

```python
#ploting the accuracy of the train and val data against epoch

plt.plot(history3.history["loss"])
plt.plot(history3.history["val_loss"])
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.legend(["Train","Validation"])
plt.show()
```
