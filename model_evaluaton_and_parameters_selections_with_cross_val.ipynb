# AIM
Step by step process of buiding the wrapper for a keras deep learning model so that it can be used in a scikit-learn workflow.

# What to do.

In order to use scikit-learn on keras,It involves building a function that will return the keras model and function will then be passed into the keras wrapper function and usinfg cross_val_score from sklearn library

# Data and type of problem
The dataset is of 908 data points 0f a regression problem where each record describes six attributes of a chemical and target is the acute toxicity toward the fish pimephales promelas.

## IN THE END::
cross_val_score will be used to select best model out of three model,and also use to pick the best hyperparameters for that model

```python
# libraries

import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,ELU,ReLU
```

```python
# loading the dataset

colnames=["CICO","SM1_Dz(Z)","GATS1i","NdscH","NdssC","MLOGP","LC50"]

data=pd.read_csv("/content/qsar_fish_toxicity.csv",names=colnames,sep=";")
```

```python
# reading the dataset
data.head()
```

```python
# spliting the data into feature and target

X=data.drop("LC50",axis=1)
Y=data[["LC50"]]
```

```python
# shape of the dataset
X.shape
```

```python
# function that will return to keras model

#The model to return will have a single hidden layer with 8 nodes using relu activation function in the hidden layer.MSE will be used as the metrics and adam optimizer for updating the parameters

def builder():
  model=Sequential()
  model.add(Dense(input_dim=X.shape[1],units=8,activation="relu"))
  model.add(Dense(units=1,activation="relu") )
  model.compile(metrics=["mean_squared_error"],optimizer="adam",loss="mean_squared_error")
  
  return model
```

```python
# scikit-Learn interface for the keras model

from keras.wrappers.scikit_learn import KerasRegressor

from sklearn.model_selection import KFold,cross_val_score

my_model=KerasRegressor(builder,epochs=100,batch_size=20)

kf=KFold(5)

result=cross_val_score(my_model,X,Y,cv=kf)
```

```python
print(result)
```

```python
print(abs(result).mean())
```

```python
# building three different model and picking the one with least error

def builder1():
  model1=Sequential()
  
  model1.add(Dense(input_dim=X.shape[1],units=4,activation="relu"))
  model1.add(Dense(units=1,activation="relu") )
  model1.compile(metrics=["mean_squared_error"],optimizer="adam",loss="mean_squared_error")
  
  return model1


def builder2():
  model2=Sequential()
  
  model2.add(Dense(input_dim=X.shape[1],units=8,activation="relu"))
  model2.add(Dense(units=1,activation="relu") )
  model2.compile(metrics=["mean_squared_error"],optimizer="adam",loss="mean_squared_error")
  
  return model2


def builder3():
  model3=Sequential()
  
  model3.add(Dense(input_dim=X.shape[1],units=4,activation="relu"))
  model3.add(Dense(units=2,activation="relu"))
  model3.add(Dense(units=1,activation="relu") )
  model3.compile(metrics=["mean_squared_error"],optimizer="adam",loss="mean_squared_error")
  
  return model3
```

```python

results=[]
models=[builder1,builder2,builder3]
for model in models:
  model=KerasRegressor(model,batch_size=20,epochs=100)
  kf=KFold(3)
  result=cross_val_score(model,X,Y,cv=kf)
  results.append(abs(result).mean())

print(f"The model with the least error is model{results.index(min(results))+1} with error {min(results)}")
```

```python
# print out the best model
print(f"The model with the least error is model{results.index(min(results))+1} with error {min(results)}")
```

```python
#using cross_val to determine  the number of epouchs and batch size for the model with the lowest error

epochs=[100,200]
batches=[20,15]
results=[]
kf=KFold(3)
for i in batches:
  for j in epochs:
    my_model=KerasRegressor(builder2,batch_size=i,epochs=j,verbose=0)
    result=cross_val_score(my_model,X,Y, cv=kf)
    results.append({"epochs":j,"batches":i,"mean_loss":abs(result).mean()})
```

```python
#the result of the above code 

results
```

# From the output,
model with batch size 20 and epoch 100 performs better

```python
# cross val score for best optimizer and activation function for the hidden layers
kf=KFold(3)
ac=["relu","elu","tanh"]
optimizers=["adam","sgd","rmsprop","adagrad"]
results=[]
for i in ac:
  for j in optimizers:
    def builder2():
  
      model2=Sequential()
        
      model2.add(Dense(input_dim=X.shape[1],units=8,activation=i))
      model2.add(Dense(units=1,activation="relu"))
      model2.compile(metrics=["mean_squared_error"],optimizer=j,loss="mean_squared_error")
    
      return model2
    my_model=KerasRegressor(builder2,batch_size=20,epochs=10,verbose=0)
    result=cross_val_score(my_model,X,Y, cv=kf)
    results.append({"optimizer":j,"ac_func":i,"mean_loss":abs(result).mean()})
```
